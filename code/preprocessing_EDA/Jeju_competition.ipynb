- 대회 직후 공개되는 Private Score 랭킹은 최종 순위가 아니며 코드 검증 후 최종 수상자가 결정됨

--> 이 의문에 따라 이용건수 = CNT , 평균 이용금액 = AMT / CNT 를 두가지 종속변수를 고려할지 결정할 수 있다.

--> 만약 두 종속변수를 고려한다면 2개의 univariate regression 모델을 만들거나 ( CNT, 평균이용금액이 상관성이 낮을 때 가능) 혹은 Multivariate regression (두 종속변수가 상관성이 있을 때)

--> 만약 CNT를 따로 모델링해도 된다면 Linear Reg는 안된다! (0 이상 count data = 정수) : Poission reg 혹은 Negbin reg 적용!

--> $\sum_{i=1}^{각 그룹에서의 사람 수}  (Cnt_i \cdot Avg_i) = 예측값$
